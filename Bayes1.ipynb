```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, poisson, binom, uniform
```

```python
params = {'a_min': 75,
          'a_max': 90,
          'b_min': 500,
          'b_max': 600,
          'p1': .1,
          'p2': .01,
          'p3': .3}
```

```python
#http://stackoverflow.com/a/1235363/2069858
def cartesian(arrays, out=None):
    arrays = [np.asarray(x) for x in arrays]
    dtype = arrays[0].dtype

    n = np.prod([x.size for x in arrays])
    if out is None:
        out = np.zeros([n, len(arrays)], dtype=dtype)

    m = n / arrays[0].size
    out[:,0] = np.repeat(arrays[0], m)
    if arrays[1:]:
        cartesian(arrays[1:], out=out[0:m,1:])
        for j in xrange(1, arrays[0].size):
            out[j*m:(j+1)*m,1:] = out[0:m,1:]
    return out
```

```python
def pc_ab (a, b, params, model=2):
    k = np.arange(40)
    p1 = params['p1']
    p2 = params['p2']
    colors = matplotlib.rcParams['axes.color_cycle'] 
    plt.figure(figsize=(12,8))

    plt.plot(k ,poisson.pmf(k, p1*a + p2*b), '-o', label='$\mathbb{E}=$%s' %(p1*a + p2*b), color=colors[0])
    plt.fill_between(k, poisson.pmf(k, p1*a + p2*b), color=colors[0], alpha=0.5)
    plt.legend()
    
    plt.title("Poisson distribution $p(c|a,b)$")
    plt.ylabel("PDF at $c$")
    plt.xlabel("$c$")

pc_ab(82, 550, {'p1': p1, 'p2': p2})
```

```python
def pd_c(c, params, model=2):
    k = np.arange(40)
    p3 = params['p3']
    rv = binom(c, p3)
    plt.figure(figsize=(10,8))
    plt.plot(k, rv.pmf(k), lw=2, color=colors[0], label='$c + c*p_3=%s$' %(c + c*p3))
    
    plt.fill_between(k, rv.pmf(k), color=colors[0], alpha=0.5)
    plt.legend()
    plt.title("Binomial distribution $p(d|c)$")
    plt.tight_layout()
    plt.ylabel("PDF at $d$")
    plt.xlabel("$d$")
    
pd_c(13.7, params)
```
